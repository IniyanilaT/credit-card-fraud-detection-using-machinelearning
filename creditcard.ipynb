```python
X = df.drop('Class',axis=1)
y = df['Class']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42,stratify=y)
```

```python
X_train.shape,X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
Scaler = StandardScaler()
X_train = Scaler.fit_transform(X_train)
X_test = Scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=100)
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
confusion_matrix(y_test,y_pred)
classification_report(y_test,y_pred)
```

```python
y_prob = model.predict_proba(X_test)
```

```python
y_prob[:5]
```

```python
y_pred_default = (y_prob[:,1]>=0.5).astype(int)
```

```python
confusion_matrix(y_test,y_pred_default)
classification_report(y_test,y_pred_default)
```

```python
threshold = 0.3
y_pred_custom = (y_prob[:,1] >= threshold).astype(int)
fraud_prob = model.predict_proba(X_test)[:,1]
```

```python
from sklearn.metrics import precision_score,recall_score
thresholds = [0.9,0.7,0.5,0.3,0.1]
for t in thresholds:
    y_pred_t = (fraud_prob >= t).astype(int)
    precision = precision_score(y_test,y_pred_t)
    recall = recall_score(y_test,y_pred_t)
    print(f"threshold:{t}")
    print(f"precision:{precision:.3f}")
    print(f"recall:{recall:.3f}")
    print("-"*30)
    
```

```python
from sklearn.metrics import confusion_matrix,classification_report
threshold = 0.3
y_pred_03 = (fraud_prob >= threshold).astype(int)
print(confusion_matrix(y_test,y_pred_03))
print(classification_report(y_test,y_pred_03))      
```

```python
from sklearn.linear_model import LogisticRegression

model_bal = LogisticRegression(
    max_iter=1000,
    class_weight='balanced'
)

model_bal.fit(X_train, y_train)
```

```python
fraud_prob_bal = model_bal.predict_proba(X_test)[:,1]
```

```python
from sklearn.metrics import confusion_matrix, classification_report

y_pred_bal = (fraud_prob_bal >= 0.5).astype(int)

print(confusion_matrix(y_test, y_pred_bal))
print(classification_report(y_test, y_pred_bal))
```

```python
threshold = 0.3
y_pred_bal_03 = (fraud_prob_bal >= threshold).astype(int)

print(confusion_matrix(y_test, y_pred_bal_03))
print(classification_report(y_test, y_pred_bal_03))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    class_weight='balanced',
    n_jobs=-1
)

rf.fit(X_train, y_train)
```

```python
fraud_prob_rf = rf.predict_proba(X_test)[:, 1]
```

```python
from sklearn.metrics import confusion_matrix, classification_report

y_pred_rf_05 = (fraud_prob_rf >= 0.5).astype(int)

print(confusion_matrix(y_test, y_pred_rf_05))
print(classification_report(y_test, y_pred_rf_05))
```

```python
threshold = 0.3
y_pred_rf_03 = (fraud_prob_rf >= threshold).astype(int)

print(confusion_matrix(y_test, y_pred_rf_03))
print(classification_report(y_test, y_pred_rf_03))
```
